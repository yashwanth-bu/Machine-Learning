```python
# Step 1: Importing the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
# Step 2: Creating a simple dataset
data = {
    'Size' : [500, 600, 700, 800, 900, 1000, 1100, 1200, 1300, 1400],
    'Price' : [150, 180, 210, 230, 250, 270, 290, 310, 330, 350]
}

df = pd.DataFrame(data)
df
```

```python
#Dataset : Size (Feature - X) and Price (Target - Y)
X = df[['Size']]
Y = df['Price']
```

```python
# Plotting the dataset
plt.scatter(X, Y, color='red', label='Actual values') 
plt.title('Simple - Linear Regression')
plt.xlabel('Size of House (sq ft)')
plt.ylabel('Price (in $1000)')
plt.legend()
plt.show()
```

```python
# Step 3: Split the data into training and testing sets (80% train, 20% test)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# Step 4: Create a linear regression model and fit it to the training data
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
# Step 5: Make predictions on the test data
Y_pred = model.predict(X_test)
```

```python
# Step 6: Evaluate the model

# R-squared value
r2 = r2_score(Y_test, Y_pred)
# Mean squared error
mse = mean_squared_error(Y_test, Y_pred)
```

```python
# Output the results
print("Linear Regression Model Coefficients:")
print(f"Intercept (β0): {model.intercept_}")
print(f"Slope (β1): {model.coef_[0]}")
print(f"R-squared: {r2}")
print(f"Mean Squared Error: {mse}")
```

```python
#Comparing the Actual and Predicated value

compareTable = pd.DataFrame({"Actual values": Y_test, 'Predicated Values' : Y_pred})
compareTable
```

```python
# Plotting the results
plt.scatter(X, Y, color='red', label='Actual values')  # Actual points
plt.plot(X, model.predict(X), color='blue', label='Regression Line')  # Predicted line
plt.title('Linear Regression: House Price Prediction')
plt.xlabel('Size of House (sq ft)')
plt.ylabel('Price (in $1000)')
plt.legend()
plt.show()
```
